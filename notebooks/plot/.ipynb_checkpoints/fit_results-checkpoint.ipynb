```python
# libraries
import sys
sys.path.append("../models/")
from vaccine_behaviour import integrate_BV
from functions import import_country, get_totR, update_contacts
import pandas as pd 
import numpy as np 
import os 
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from functions import get_beta, update_contacts
import seaborn as sns
from matplotlib.ticker import FormatStrFormatter

import matplotlib.font_manager as font_manager
import matplotlib.patches as patches

plt.rcParams['axes.linewidth'] = 0.3
plt.rcParams['xtick.major.width'] = 0.3
plt.rcParams['ytick.major.width'] = 0.3
plt.rcParams['xtick.major.size'] = 3
plt.rcParams['ytick.major.size'] = 3
plt.rcParams['xtick.minor.width'] = 0.2
plt.rcParams['ytick.minor.width'] = 0.2
plt.rcParams['xtick.minor.size'] = 1.5
plt.rcParams['ytick.minor.size'] = 1.5

font_dirs = ['./fonts/Bai_Jamjuree/']
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)
font_list = font_manager.createFontList(font_files)
font_manager.fontManager.ttflist.extend(font_list)

plt.rcParams['font.family'] = 'Bai Jamjuree'
plt.rcParams['font.sans-serif'] = 'Bai Jamjuree'

# datetimes
start_date = datetime(2020, 8, 31)
end_date = datetime(2021, 1, 4)

# error metric
def wmape(arr1, arr2):
    # weigthed mape
    return np.sum(np.abs(arr1 - arr2)) / np.sum(np.abs(arr1))
```

```python
ncomp = 12
nage  = 16

colors = ['#6CC2BD', '#5A809E', '#7C79A2', '#F57D7C', '#FFC1A6', '#FEE4C4']

# add week of year
dates = [datetime(2020, 8, 31) + timedelta(days=d) for d in range(126)]
year_week = []
for date in dates: 
    y, w = date.isocalendar()[0], date.isocalendar()[1]
    if w < 9: 
        year_week.append(str(y) + "-0" + str(w))
    else: 
        year_week.append(str(y) + "-" + str(w))
        
        
fig, axes = plt.subplots(nrows=3, ncols=2, dpi=210, figsize=(10,7))

i, j = 0, 0
xs = np.unique(year_week)
xs = [x.replace("2020-", "") for x in xs]

for country in ["Egypt", "Peru", "Serbia", "Ukraine", "Canada", "Italy"]:
    
    # import country
    country_dict = import_country(country, "../../data/countries/")
    
    # deaths real 
    deaths_real = country_dict["deaths"].loc[(country_dict["deaths"]["year_week"]>="2020-36") & \
                                             (country_dict["deaths"]["year_week"]<="2020-53")]["weekly_count"].values

    params = []
    files  = os.listdir("../output/posterior/")
    files  = [file for file in files if country in file]

    params_sampled = []
    for file in files:
        if params_sampled == []:
            params_sampled = np.load("../output/posterior/" + file, allow_pickle=True)["arr_0"]
        else: 
            params = np.load("../output/posterior/" + file, allow_pickle=True)["arr_0"]
            params_sampled = np.concatenate((params_sampled, params))

    R0_sampled     = [p[0] for p in params_sampled] 
    Delta_sampled  = [p[1] for p in params_sampled]
    i0_sampled     = [p[2] for p in params_sampled] 
    deaths_sampled = [p[3] for p in params_sampled] 
    
    print(len(R0_sampled))
    axes[i, j].grid(color='grey', linestyle='--', linewidth=0.35, alpha=0.6, axis="both")
    axes[i, j].spines['top'].set_visible(False)
    axes[i, j].spines['right'].set_visible(False)
    axes[i, j].minorticks_on()  

    p1, = axes[i, j].plot(xs[2:], np.median(deaths_sampled, axis=0)[2:], color="#6CC2BD")
    p2  = axes[i, j].fill_between(xs[2:], 
                         np.quantile(deaths_sampled, axis=0, q=0.025)[2:], 
                         np.quantile(deaths_sampled, axis=0, q=0.975)[2:], color="#6CC2BD", alpha=0.1, linewidth=0)

    axes[i, j].fill_between(xs[2:], 
                         np.quantile(deaths_sampled, axis=0, q=0.25)[2:], 
                         np.quantile(deaths_sampled, axis=0, q=0.75)[2:], color="#6CC2BD", alpha=0.3, linewidth=0)#6CC2BD

    p3 = axes[i, j].scatter(xs[2:], deaths_real[2:], color="#5A809E", edgecolor="black", linewidth=0.2, s=9)
    axes[i, j].set_title(country.replace("_", ""))
    
    axes[i, j].tick_params(axis="x", rotation=0, labelsize=8)
    axes[i, j].set_ylabel("weekly deahts")
    axes[i, j].set_xlabel("week (2020)", fontsize=8)
    j += 1
    
    if j==2:
        j = 0
        i += 1
        
axes[2, 0].legend([(p1,p2), p3], ["median (50%, 95% CI)", "observed"],
                       bbox_to_anchor=[0.5, -1.4,1,1], frameon=False, ncol=2)
plt.subplots_adjust(hspace=0.55, wspace=0.2)
plt.savefig("./img/fit_results.pdf", bbox_inches="tight")
```

```python
countries = ['Italy', 'Canada', 'Ukraine', 'Serbia', 'Peru', 'Egypt']
colors = ['#6CC2BD', '#5A809E', '#F57D7C', '#7C79A2', '#FEE4C4', '#FFC1A6']
eps   = 1.0 / 3.7
mu    = 1.0 / 2.5
omega = 1.0 / 1.5
chi   = 0.55
f     = 0.35 
fig, axes = plt.subplots(ncols=3, dpi=300, figsize=(10,3))

i = 0   
for country in countries:
    
    country_dict = import_country(country, "../../data/countries/")
    
    params = []
    files  = os.listdir("../output/posterior/")
    files  = [file for file in files if country in file]

    params_sampled = []
    for file in files:
        if params_sampled == []:
            params_sampled = np.load("../output/posterior/" + file, allow_pickle=True)["arr_0"]
        else: 
            params = np.load("../output/posterior/" + file, allow_pickle=True)["arr_0"]
            params_sampled = np.concatenate((params_sampled, params))

    R0_sampled     = np.array([p[0] for p in params_sampled]) 
    Delta_sampled  = np.array([p[1] for p in params_sampled])
    i0_sampled     = np.array([p[2] for p in params_sampled]) 
    deaths_sampled = np.array([p[3] for p in params_sampled]) 
    C = update_contacts(country_dict, datetime(2020, 9, 1))
    betas = np.array([get_beta(R0, mu, chi, omega, f, C, country_dict["Nk"]) for R0 in R0_sampled])
    
    initial_I = i0_sampled * 100000
    Q1, Q3 = np.quantile(initial_I, q=0.25), np.quantile(initial_I, q=0.75)
    IQR    = Q3 - Q1
    axes[1].plot([Q1, Q3], [i,i], linewidth=10, color=colors[i], alpha=0.5)
    axes[1].plot([Q1 - 1.5 * IQR, Q3 + 1.5 * IQR], [i,i], linewidth=10, color=colors[i], alpha=0.3)
    axes[1].scatter([np.median(initial_I)], [i], color=colors[i], marker="|", s=200, linewidth=3)
    
    Q1, Q3 = np.quantile(betas, q=0.25), np.quantile(betas, q=0.75)
    IQR    = Q3 - Q1
    axes[0].plot([Q1, Q3], [i,i], linewidth=10, color=colors[i], alpha=0.5)
    axes[0].plot([Q1 - 1.5 * IQR, Q3 + 1.5 * IQR], [i,i], linewidth=10, color=colors[i], alpha=0.3)
    axes[0].scatter([np.median(betas)], [i], color=colors[i], marker="|", s=200, linewidth=3)
    
    
    Q1, Q3 = np.quantile(Delta_sampled, q=0.25), np.quantile(Delta_sampled, q=0.75)
    IQR    = Q3 - Q1
    axes[2].plot([Q1, Q3], [i,i], linewidth=10, color=colors[i], alpha=0.5)
    axes[2].plot([Q1 - 1.5 * IQR, Q3 + 1.5 * IQR], [i,i], linewidth=10, color=colors[i], alpha=0.3)
    axes[2].scatter([np.median(Delta_sampled)], [i], color=colors[i], marker="|", s=200, linewidth=3)
    i += 1 
    
axes[0].set_title("$\\beta$")
axes[1].set_title("Initial Infected per 100'000")
axes[2].set_title("$\\Delta$")

for ax in axes:
    ax.set_ylim(-0.5, 5.5)
    ax.set_yticks([0,1,2,3,4,5])
    ax.set_yticklabels(["", "", "", "", "", ""])
    ax.grid(color='grey', linestyle='--', linewidth=0.35, alpha=0.6, axis="both")
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
axes[0].set_yticklabels(countries)
plt.savefig("./img/posteriors.pdf", bbox_inches="tight")
```
