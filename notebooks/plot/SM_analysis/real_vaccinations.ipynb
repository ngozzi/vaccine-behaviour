```python
def import_boxplot_real(country, cases=False):
    
    vaccination_strategies = ["real"]

    files = os.listdir("../../output/real_runs/")
    files = [file for file in files if country in file]
    files = [file for file in files if "real" in file]
    
    if cases: 
        files = [file for file in files if "cases" in file]
    else:
        files = [file for file in files if "cases" not in file]
           
    # initialize solutions
    data          = dict()
    data[country] = dict()
    for vaccination_strategy in vaccination_strategies:
        data[country][vaccination_strategy] = dict()
        for r in rs:
            data[country][vaccination_strategy][r] = dict()
            for alpha in alpha_s:
                data[country][vaccination_strategy][r][alpha] = []

    for file in files:
        with open("../../output/real_runs/" + file, "rb") as file:  
            data_temp = pkl.load(file)
            for vaccination_strategy in vaccination_strategies:
                for r in rs:
                    for alpha in alpha_s:
                        for el in data_temp[country][vaccination_strategy][r][alpha]:
                            data[country][vaccination_strategy][r][alpha].append(el)
                            
    return data


bp_ita_deaths = import_boxplot_real("Italy", False)
```

```python
import pickle as pkl
with open("../../../data/countries/Italy/vaccinations/vaccinations.pkl", "rb") as file:
    vaccinations = pkl.load(file)
    
country_dict = import_country("Italy", "../../../data/countries/")
dates, daily_vacc, vacc_by_age = [], [],  np.zeros(16)

for k in vaccinations.keys():
    dates.append(k)
    daily_vacc.append(100 * np.sum(vaccinations[k]) / country_dict["Nk"].sum())
    vacc_by_age += vaccinations[k]
```

```python
import matplotlib.gridspec as gridspec
fig = plt.figure(constrained_layout=True,  dpi=500, figsize=(10,7))
spec = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)

ax1 = fig.add_subplot(spec[0, :])
ax2 = fig.add_subplot(spec[1, 0])
ax3 = fig.add_subplot(spec[1, 1])

ax1.plot(range(len(alpha_s)), 
             [np.median(bp_ita_deaths["Italy"]["real"][1.5][alpha]) for alpha in alpha_s], 
             color=colors[0], marker="o", label="$r=1.5$", markersize=10)
ax1.plot(range(len(alpha_s)), 
             [np.median(bp_ita_deaths["Italy"]["real"][1.3][alpha]) for alpha in alpha_s], 
             color=colors[1], marker="s", label="$r=1.3$", markersize=10)

ax1.set_title("Italy - Real Vaccination (up to 2021/06/01)", fontsize=15)
ax1.set_ylabel("Relative deaths difference")
ax1.set_xticks([0,1,2,3,4])
ax1.set_xticklabels(["0\nbaseline","$10^{-1}$","$10^{0}$","$10^{1}$","$10^{2}$"])
ax1.set_xlabel("$\\alpha$", fontsize=15)
ax1.legend(loc="lower left", frameon=False)
ax1.hlines(0, -0.9, 4.5, linestyle="--", color="gray")
ax1.set_xlim(xmin=-0.1, xmax=4.1)
ax1.tick_params(axis="x", labelsize=12)


for ax in [ax1, ax2, ax3]:
    ax.grid(color='grey', linestyle='--', linewidth=0.35, alpha=0.6, axis="both")
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    
ax2.fill_between(dates, 0, np.cumsum(daily_vacc), color=colors[0], alpha=0.5, linewidth=0)
ax2.plot(dates, np.cumsum(daily_vacc), color=colors[0], linewidth=2)
ax2.set_title("Population with at least 1 dose (%)", y=1.1)
ax2.tick_params(axis="x", rotation=30)
#plt.tight_layout()

ax3.bar(range(16), country_dict["Nk"], alpha=0.5,color=colors[0], label="Total number of people", edgecolor="k", linewidth=0.5)
ax3.bar(range(16), vacc_by_age, hatch="xx", color=colors[0], label="People with at least one dose", edgecolor=colors[1])
ax3.set_title("People with at least 1 dose in different age groups", y=1.1)
ax3.legend(loc="upper left", frameon=False)

i = 0
for n in country_dict["Nk"]:
    if 100 * vacc_by_age[i] / country_dict["Nk"][i] < 0.1:
        t = "< 0.1"
    else:
        t = np.round(100 * vacc_by_age[i] / country_dict["Nk"][i], 1)
    ax3.text(i, n+0.05*n, str(t), ha="center", size=8)
    i += 1
ax3.set_xticks(range(16))
ax3.set_xticklabels(["0-4", "5-9", '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', 
                      '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74', "75+"], rotation=90, size=10)
ax3.tick_params(axis="y", labelsize=10)

plt.savefig("./SM_img/real_vaccinations_Italy.pdf")
```
