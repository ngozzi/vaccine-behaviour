```python
import matplotlib.pyplot as plt 
import numpy as np 
import matplotlib.font_manager as font_manager
from matplotlib import cm
import matplotlib.lines as lines
colors = ['#6CC2BD', '#5A809E', '#7C79A2', '#F57D7C', '#FFC1A6', '#FEE4C4']

plt.rcParams['axes.linewidth'] = 0.3
plt.rcParams['xtick.major.width'] = 0.3
plt.rcParams['ytick.major.width'] = 0.3
plt.rcParams['xtick.major.size'] = 3
plt.rcParams['ytick.major.size'] = 3
plt.rcParams['xtick.minor.width'] = 0.2
plt.rcParams['ytick.minor.width'] = 0.2
plt.rcParams['xtick.minor.size'] = 1.5
plt.rcParams['ytick.minor.size'] = 1.5

font_dirs = ['../fonts/Bai_Jamjuree/']
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)
font_list = font_manager.createFontList(font_files)
font_manager.fontManager.ttflist.extend(font_list)
plt.rcParams['font.family'] = 'Bai Jamjuree'
plt.rcParams['font.sans-serif'] = 'Bai Jamjuree'

def g(alpha, vt):
    return (1 - np.exp(-alpha * vt))
```

```python
from matplotlib import cm

fig, axes = plt.subplots(nrows=2, figsize=(5,4), dpi=300)

ax = axes[0]
cmap = cm.get_cmap('GnBu')
vts = np.linspace(0, 1, 500)
alphas = np.array([100, 10, 5, 3, 2, 1, 0.5, 0.3, 0.1])

for i in range(len(alphas)):
    ax.plot(vts, g(alphas[i], vts), label="$\\alpha=$" + str(alphas[i]), color=cmap((10 - i) / 10))

ax.legend(frameon=False, ncol=1, loc="upper right", bbox_to_anchor=[1.2,1,0,0], fontsize=5)
ax.set_xlabel("Fraction of the population vaccinated ($v_t$)", fontsize=8)
ax.set_ylabel("$\\lambda_{X \\rightarrow X^{NC}}$")
ax.tick_params(labelsize=6)

ax.hlines(0.5, -1, 2, linestyle="--", color="gray")
ax.set_xlim(-0.01,1.01)
ax.grid(color='grey', linestyle='--', linewidth=0.35, alpha=0.6, axis="both")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax = axes[1]
cmap = cm.get_cmap('YlOrBr')
dts = np.linspace(0, 5, 500)
gammas = np.array([100, 10, 5, 3, 2, 1, 0.5, 0.3, 0.1])

for i in range(len(gammas)):
    ax.plot(dts, g(gammas[i], dts), label="$\\gamma=$" + str(gammas[i]), color=cmap((10 - i) / 10))

ax.legend(frameon=False, ncol=1, loc="upper right", bbox_to_anchor=[1.2,1,0,0], fontsize=5)
ax.set_xlabel("Daily deaths per 100'000 ($d^{o}_t$)", fontsize=8)
ax.set_ylabel("$\\lambda_{X^{NC} \\rightarrow X}$")

ax.hlines(0.5, -1, 6, linestyle="--", color="gray")
ax.set_xlim(-0.05, 5.05)
ax.grid(color='grey', linestyle='--', linewidth=0.35, alpha=0.6, axis="both")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(labelsize=6)

fig.suptitle("Behavioural Transitions Rates", y=0.96, fontsize=10)
plt.subplots_adjust(hspace=0.5)

fig.add_artist(lines.Line2D([0.05, 1.05], [0.465, 0.465], linestyle="dotted", color="gray", linewidth=1))

axes[0].annotate("A", xy=(-0.1, 1.), xycoords='axes fraction', fontsize=10, va='center', ha='left', weight="bold")
axes[1].annotate("B", xy=(-0.1, 1.), xycoords='axes fraction', fontsize=10, va='center', ha='left', weight="bold")

plt.savefig("./SM_img/behavioural_rates.pdf", bbox_inches="tight")
```

```python
# libraries
import sys
sys.path.append("../../models/")
from vaccine_behaviour import integrate_BV
from functions import import_country, get_totR, update_contacts
import pandas as pd 
import numpy as np 
import os 
from datetime import datetime, timedelta

# epi params
eps   = 1.0 / 3.7
mu    = 1.0 / 2.5
omega = 1.0 / 1.5
chi   = 0.55
f     = 0.35 
IFR   = [0.00161 / 100, # 0-4  
         0.00161 / 100, # 5-9
         0.00695 / 100, # 10-14
         0.00695 / 100, # 15-19 
         0.0309  / 100, # 20-24 
         0.0309  / 100, # 25-29
         0.0844  / 100, # 30-34
         0.0844  / 100, # 35-39
         0.161   / 100, # 40-44 
         0.161   / 100, # 45-49 
         0.595   / 100, # 50-54 
         0.595   / 100, # 55-59 
         1.93    / 100, # 60-64
         1.93    / 100, # 65-69 
         4.28    / 100, # 70-74 
         6.04    / 100] # 75+
R0      = 1.15
Delta   = 17
i0      = 1.0 / 100.0
r0      = 10.0 / 100.0

start_date = datetime(2021, 1, 1)
end_date   = datetime(2022, 1, 1)

VES = 0.7
VEM = 0.6

country_dict = import_country("Italy", "../../../data/countries/")
C = country_dict["contacts_home"] + country_dict["contacts_work"] + \
    country_dict["contacts_school"] + country_dict["contacts_other_locations"]
Nk = country_dict["Nk"]

Cs = {}
date, dates = start_date, [start_date]
for i in range((end_date - start_date).days): 
    Cs[date] = C
    date += timedelta(days=1)
    dates.append(date)
    
ics = np.zeros((16, 20))
for k in range(16):
    ics[k, 4] = eps**(-1) / (mu**(-1) + omega**(-1) + eps**(-1)) * i0 * country_dict["Nk"][k]
    ics[k, 5] = omega**(-1) / (mu**(-1) + omega**(-1) + eps**(-1)) * i0 * country_dict["Nk"][k]
    ics[k, 6] = f * mu**(-1) / (mu**(-1) + omega**(-1) + eps**(-1)) * i0 * country_dict["Nk"][k]
    ics[k, 7] = (1 - f) * mu**(-1) / (mu**(-1) + omega**(-1) + eps**(-1)) * i0 * country_dict["Nk"][k]
    ics[k, 9] = r0 * country_dict["Nk"][k]
    ics[k, 0] = country_dict["Nk"][k] - ics[k, 4] - ics[k, 5] - ics[k, 6] - ics[k, 7] - ics[k, 9]
```

```python
fig, axes = plt.subplot_mosaic([["A", "B", "C"],
                    ["A", "B", "C"], 
                    ["D", "E", "F"], 
                    ["G", "H", "I"]], figsize=(10,7), dpi=300)

letters = {1.15: ["A", "D", "G"], 1.5: ["B", "E", "H"], 2.0: ["C", "F", "I"]}
for R0 in [1.15, 1.5, 2.0]:
    
    z = 0 
    for alpha in [0.1, 1, 10]:

        sol, Vt, vs = integrate_BV(ics, (end_date - start_date).days, R0=R0, eps=eps, mu=mu, 
                                   omega=omega, chi=chi, f=f, IFR=IFR, Delta=Delta, r=1.3, 
                                   alpha=alpha, gamma=0.5, rV=0.25 / 100, VES=VES, VEM=VEM, Cs=Cs, 
                                   Nk=country_dict["Nk"], vaccination_strategy="homogeneous", 
                                   model="vaccine_rate", dates=dates)

        sol = sol.sum(axis=0)
        NC = (sol[2, :] + sol[3, :]) / Nk.sum()
        DO = 100000 * np.diff(sol[11, :] + sol[19, :]) /  np.sum(Nk) 
        V = np.cumsum(Vt) / Nk.sum()
        
        
        axes[letters[R0][0]].plot(NC, color=colors[z], label="$\\alpha=$" + str(alpha))
        axes[letters[R0][1]].plot(V, color=colors[z])
        axes[letters[R0][2]].plot(DO, color=colors[z])
        
        z += 1

for ax in axes.values(): 
    ax.grid(color='grey', linestyle='--', linewidth=0.35, alpha=0.6, axis="both")
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
axes["A"].legend(frameon=False, loc="lower right")
axes["A"].set_ylim(0, 0.95)
axes["B"].set_ylim(0, 0.95)
axes["C"].set_ylim(0, 0.95)

axes["D"].set_ylim(0, 0.85)
axes["E"].set_ylim(0, 0.85)
axes["F"].set_ylim(0, 0.85)

axes["G"].set_ylim(0, 6.5)
axes["H"].set_ylim(0, 6.5)
axes["I"].set_ylim(0, 6.5)

axes["A"].set_ylabel("fraction of NC")
axes["D"].set_ylabel("$v_t$")
axes["G"].set_ylabel("Deaths per 100'000")

axes["G"].set_xlabel("$t$")
axes["H"].set_xlabel("$t$")
axes["I"].set_xlabel("$t$")

axes["A"].set_title("$R_0 = 1.15$")
axes["B"].set_title("$R_0 = 1.5$")
axes["C"].set_title("$R_0 = 2.0$")

plt.tight_layout()

plt.savefig("./SM_img/nc_time.pdf", bbox_inches="tight")
```
